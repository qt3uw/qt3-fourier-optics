```python
# Imports for data analysis, no instrument control.  STUDENTS CAN RUN THESE METHODS FROM HOME
import matplotlib.pyplot as plt
import numpy as np
from lab_utilities import save_image, load_image, fourier_transform_image, azimuthal_average
```

```python
# Imports for camera acquisition, WILL FAIL IF NOT ON LAB COMPUTER
from thorcam import ThorCam
cam = ThorCam('26734', default_timeout_s=15.0) # Sometimes (maybe with wrong USB port), camera loading takes 13 s.  That problem seems to have gone away though.
cam.open()
```

```python
cam.set_exposure_ms(15.) # Set exposure to 15 ms
cam.set_gain(5) # Set gain to 5 dB
name = "example" # Choose a name for saving your image, a unique timestamp will be appended.  The saved image will be in the local data/ folder
img = cam.snap()

plt.hist(img.flatten())

fig, ax = plt.subplots(1, 1)  # Generating fig, ax objects gives you easy control later over formatting of titles, etc. - strongly recommended
ax.imshow(img)
print(f'original data type: {img.dtype}')
print(f'original shape: {img.shape}')
ax.set_title(name)
first_image_path = save_image(img, name=name)
```
