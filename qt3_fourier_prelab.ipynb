```python
# Imports for data analysis, no instrument control.  STUDENTS CAN RUN THESE METHODS FROM HOME
import matplotlib.pyplot as plt
import numpy as np
from lab_utilities import save_image, load_image, fourier_transform_image, azimuthal_average, load_grace_hopper, load_stripes
```

```python
img_gh = load_grace_hopper() # Natural image, a portrait of Grace Hopper
img_stripes = load_stripes(orientation="vertical", bright_width=2, dark_width=1) # A binary stripe pattern, periodic.
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_gh, cmap='gray')
ax[0].set_title('Grace Hopper')

## Plot grace hoppers fourier transform, using the fourier_transform_image method we imported above from lab_utilities.py.  
# Try plotting it both in linear units and in logarithmic (decibel) units.  You will see logarthimic units are much better for visualizing
# the fourier transform, largely because most of the energy is usually concentrated in the low spatial frequencies, and the high spatial 
# frequencies are much weaker.

ax[1].set_title('Grace Hopper Fourier Transform')
```

```python
# Do the same thing for the stripes image.  You should see that the fourier transform of the stripes has a very different structure 
# than the fourier transform of the natural image, and that it has a lot of energy at high spatial frequencies, 
# which is a consequence of the sharp edges in the stripes pattern.  How does the fourier transform change when you change the stripe periodicity?
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_stripes, cmap='gray')
ax[0].set_title('stripes')

ax[1].set_title('Stripes Fourier Transform')
```
